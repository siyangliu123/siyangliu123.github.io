```python
gramsList = nutritionDf['Grams'].tolist()
caloriesList = nutritionDf['Calories'].tolist()
proteinList = nutritionDf['Protein'].tolist()
fatList = nutritionDf['Fat'].tolist()
satList = nutritionDf['Sat.Fat'].tolist()
fiberList = nutritionDf['Fiber'].tolist()
carbList = nutritionDf['Carbs'].tolist()
```

```python
def numberFunc(ListName):
    for n,i in enumerate(ListName):
        if i.isdigit() == False:
            ListName[n] = 0
    return ListName


caloriesList = numberFunc(caloriesList)

fatList = numberFunc(fatList)

proteinList = numberFunc(proteinList)

satList = numberFunc(satList)

fiberList = numberFunc(fiberList)

carbList = numberFunc(carbList)
        
# for n,i in enumerate(fatList):
#     if i.isdigit() == False:
#         fatList[n] = 0   
# for n,i in enumerate(proteinList):
#     if i.isdigit() == False:
#         proteinList[n] = 0
# for n,i in enumerate(satList):
#     if i.isdigit() == False:
#         satList[n] = 0
# for n,i in enumerate(fiberList):
#     if i.isdigit() == False:
#         fiberList[n] = 0
# for n,i in enumerate(carbList):
#     if i.isdigit() == False:
#         carbList[n] = 0
```

```python
def replaceFunc(Listname):
    count = 0
    for i in Listname:
        num = str(i).replace(',', '')
        Listname[count] = num
        count += 1
    return Listname

gramsList = replaceFunc(gramsList)
caloriesList = replaceFunc(caloriesList)
fatList = replaceFunc(fatList)
proteinList = replaceFunc(proteinList)
satList = replaceFunc(satList)
fiberList = replaceFunc(fiberList)
carbList = replaceFunc(carbList)
    
```

```python
caloriesList
```

```python
newcaloriesList = []
newproteinList = []
newfatList = []
newsatList = []
newfiberList = []
newcarbList = []
num = 0
for i in gramsList:
    a= (float(caloriesList[num])*100)/float(gramsList[num])
    b= (float(proteinList[num])*100)/float(gramsList[num])
    c= (float(fatList[num])*100)/float(gramsList[num])
    d= (float(satList[num])*100)/float(gramsList[num])
    e= (float(fiberList[num])*100)/float(gramsList[num])
    f= (float(carbList[num])*100)/float(gramsList[num])
    newcaloriesList.append(a)
    newproteinList.append(b)
    newfatList.append(c)
    newsatList.append(d)
    newfiberList.append(e)
    newcarbList.append(f)
    num += 1
```

```python
newGramsList=[]
newGramsList.extend([100] * 334)
```

```python
nutritionDf['Calories'] = newcaloriesList
nutritionDf['Protein'] = newproteinList
nutritionDf['Fat'] = newfatList
nutritionDf['Sat.Fat'] = newsatList
nutritionDf['Fiber'] = newfiberList
nutritionDf['Carbs'] = newcarbList
nutritionDf['Grams'] = newGramsList
```
