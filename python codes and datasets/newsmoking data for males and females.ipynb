```python
yearList1 = smokingMaledf1['2001'].tolist()
yearList2 = smokingMaledf1['2004'].tolist()
yearList3 = smokingMaledf1['2007'].tolist()
yearList4 = smokingMaledf1['2010'].tolist()
yearList5 = smokingMaledf1['2013'].tolist()
yearList6 = smokingMaledf1['2016'].tolist()
yearList7 = smokingMaledf1['2019'].tolist()
```

```python
def replaceFunc(Listname):
    count = 0
    for i in Listname:
        num = str(i).replace(',', '')
        Listname[count] = num
        count += 1
    return Listname
```

```python
yearList1 = replaceFunc(yearList1)
yearList2 = replaceFunc(yearList2)
yearList3 = replaceFunc(yearList3)
yearList4 = replaceFunc(yearList4)
yearList5 = replaceFunc(yearList5)
yearList6 = replaceFunc(yearList6)
yearList7 = replaceFunc(yearList7)
```

```python
yearList1
```

```python
def intfunc(List):
    count = 0
    for i in List:
        i = int(i)
        List[count] = i
        count += 1
    return List
```

```python
yearList1 = intfunc(yearList1)
yearList2 = intfunc(yearList2)
yearList3 = intfunc(yearList3)
yearList4 = intfunc(yearList4)
yearList5 = intfunc(yearList5)
yearList6 = intfunc(yearList6)
yearList7 = intfunc(yearList7)
```

```python
yearList1
```

```python
smokingMaledf1['2001'] = yearList1
smokingMaledf1['2004'] = yearList2
smokingMaledf1['2007'] = yearList3
smokingMaledf1['2010'] = yearList4
smokingMaledf1['2013'] = yearList5
smokingMaledf1['2016'] = yearList6
smokingMaledf1['2019'] = yearList7
```

```python
smokingMaledf1['2001']= (smokingMaledf1['2001']/smokingMaledf1['2001'].sum())* 100
smokingMaledf1['2004']= (smokingMaledf1['2004']/smokingMaledf1['2004'].sum())* 100
smokingMaledf1['2007']= (smokingMaledf1['2007']/smokingMaledf1['2007'].sum())* 100
smokingMaledf1['2010']= (smokingMaledf1['2010']/smokingMaledf1['2010'].sum())* 100
smokingMaledf1['2013']= (smokingMaledf1['2013']/smokingMaledf1['2013'].sum())* 100
smokingMaledf1['2016']= (smokingMaledf1['2016']/smokingMaledf1['2016'].sum())* 100
smokingMaledf1['2019']= (smokingMaledf1['2019']/smokingMaledf1['2019'].sum())* 100
```
